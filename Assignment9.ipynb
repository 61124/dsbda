```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=sns.load_dataset('titanic')
```

```python
dataframe.columns
```

```python
#box plot for age wrt gender with info of whether they survived
sns.boxplot(x='sex',y='age',hue='survived',data=dataframe)
plt.title("Age vs Gender with survival information")
plt.legend(title='Survived')
plt.show()
```
