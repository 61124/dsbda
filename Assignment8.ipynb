```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import seaborn as sns

df = sns.load_dataset('titanic')  # correct name, no .csv
df.head()
```

```python
df.columns
```

```python
sns.countplot(x='sex',hue='survived',data=df)
plt.title('Survival based on gender')
plt.legend(title="Survival Rate",labels=["no","yes"])
plt.show()
```

```python
sns.countplot(x='class',hue='survived',data=df)
plt.title('Survival based on class')
plt.legend(title="Survival Rate",labels=["no","yes"])
plt.show()
```

```python
sns.histplot(x='age',hue='survived',kde=True,multiple='stack',data=df)
plt.title("Age Distirbution")
plt.show()
```

```python
sns.barplot(x='pclass',y='fare',data=df)
plt.title("Avg fare by class")
plt.show()
```

```python
sns.violinplot(x='pclass',y='age',hue='survived',data=df)
plt.xlabel("Class")
```

```python
sns.scatterplot(x='age',y='fare',hue='survived',data=df)
plt.show()
```

```python
sns.histplot(x='fare',hue='survived',data=df,multiple="stack",kde=True)
plt.show()
```
