```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=sns.load_dataset('titanic')
```

```python
dataframe.head()
```

```python
dataframe.columns
```

```python
sns.countplot(x='sex',hue='survived',data=dataframe)
plt.show()
```

```python
sns.countplot(x='pclass',hue='survived',data=dataframe)
plt.show()
```

```python
sns.histplot(x='age',hue='survived',multiple='stack',kde=True,data=dataframe)
plt.show()
```

```python
sns.barplot(x='pclass',y='fare',data=dataframe)
```

```python
sns.violinplot(x='pclass',y='age',hue='survived',data=dataframe)
```

```python
sns.scatterplot(x='age',y='fare',hue='survived',data=dataframe)
```

```python
numeric=dataframe.select_dtypes(include='number')
corr=numeric.corr()
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
sns.histplot(x='fare',kde=True,hue='survived',multiple='stack',data=dataframe)
```
