```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe=sns.load_dataset("iris")
dataframe.head()
```

```python
dataframe.columns
```

```python
# features and their types
for col in dataframe.columns:
    d=dataframe[col].dtype
    if d=='object':
        print(f"{col}: Nominal")
    else:
        print(f"{col}: Numeric")
```

```python
dataframe.drop('Id',inplace=True,axis=1)
```

```python
dataframe.columns
```

```python
numeric_cols=dataframe.select_dtypes(include=['int','float64']).columns
```

```python
# histogram for each feature
for col in numeric_cols:
    sns.histplot(x=col,bins=30,kde=True,data=dataframe)
    plt.show()
```

```python
# box plot for each feature
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=col,data=dataframe)
    plt.show()
```

```python
# finding outliers using IQR
for col in numeric_cols:
    q1=dataframe[col].quantile(0.25)
    q3=dataframe[col].quantile(0.75)
    IQR=q3-q1
    upper_bound=q3+1.5*IQR
    lower_bound=q1-1.5*IQR
    outliers=dataframe[(dataframe[col]>upper_bound)| (dataframe[col]<lower_bound)]
    print(f"{col}--{len(outliers)}")
```
